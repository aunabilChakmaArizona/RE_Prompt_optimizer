```python
import re

text = """[agent_gradient_eval_debug] prompt full evaluation: generation_index=0 beam_index=0 split=dev done in 1978.50s, precision=11.25 ± 1.89, recall=41.01 ± 10.38, f1=17.51 ± 2.82
[agent_gradient_eval_debug] prompt full evaluation: generation_index=0 beam_index=0 stable_f1=11.86 (k=2)
[agent_gradient_eval_debug] prompt full evaluation: generation_index=1 beam_index=1 split=dev done in 1972.66s, precision=11.24 ± 1.85, recall=31.97 ± 4.91, f1=16.41 ± 1.71
[agent_gradient_eval_debug] prompt full evaluation: generation_index=1 beam_index=1 stable_f1=12.99 (k=2)
[agent_gradient_eval_debug] prompt full evaluation: generation_index=2 beam_index=2 split=dev done in 1953.30s, precision=12.37 ± 3.46, recall=33.57 ± 8.40, f1=17.75 ± 3.94
[agent_gradient_eval_debug] prompt full evaluation: generation_index=2 beam_index=2 stable_f1=9.86 (k=2)
[agent_gradient_eval_debug] prompt full evaluation: generation_index=3 beam_index=3 split=dev done in 1395.98s, precision=12.30 ± 2.51, recall=30.64 ± 6.18, f1=17.29 ± 2.59
[agent_gradient_eval_debug] prompt full evaluation: generation_index=3 beam_index=3 stable_f1=12.11 (k=2)
[agent_gradient_eval_debug] prompt full evaluation: generation_index=4 beam_index=4 split=dev done in 1338.87s, precision=10.42 ± 2.17, recall=35.98 ± 6.72, f1=16.00 ± 2.71
[agent_gradient_eval_debug] prompt full evaluation: generation_index=4 beam_index=4 stable_f1=10.57 (k=2)"""

pattern = re.compile(r"f1=(\d+(?:\.\d+)?) ± (\d+(?:\.\d+)?)")

rows = []
for i, match in enumerate(pattern.finditer(text), start=1):
    mean = float(match.group(1))
    std = float(match.group(2))
    score = mean - 2.0 * std

    rows.append({
        "original_rank": i,
        "f1_mean": mean,
        "f1_std": std,
        "score": score,
    })

ranked = sorted(rows, key=lambda x: x["score"], reverse=True)

for rank, row in enumerate(ranked, start=1):
    print(
        f"{rank:2d}. line={row['original_rank']:2d} "
        f"f1={row['f1_mean']:.2f} ± {row['f1_std']:.2f} "
        f"score={row['score']:.2f}"
    )
```
