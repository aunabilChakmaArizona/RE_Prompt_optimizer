```python
import re

text = """[agent_gradient_eval_debug] prompt full evaluation: generation_index=0 beam_index=0 split=dev done in 523.29s, precision=13.20 ± 3.67, recall=27.46 ± 5.56, f1=17.63 ± 3.99
[agent_gradient_eval_debug] prompt full evaluation: generation_index=0 beam_index=0 stable_f1=9.66 (k=2)
[agent_gradient_eval_debug] prompt full evaluation: generation_index=1 beam_index=1 split=dev done in 524.29s, precision=12.78 ± 4.24, recall=27.46 ± 5.56, f1=17.08 ± 4.23
[agent_gradient_eval_debug] prompt full evaluation: generation_index=1 beam_index=1 stable_f1=8.61 (k=2)
[agent_gradient_eval_debug] prompt full evaluation: generation_index=2 beam_index=2 split=dev done in 524.28s, precision=13.10 ± 3.79, recall=27.46 ± 5.56, f1=17.46 ± 3.90
[agent_gradient_eval_debug] prompt full evaluation: generation_index=2 beam_index=2 stable_f1=9.66 (k=2)
[agent_gradient_eval_debug] prompt full evaluation: generation_index=3 beam_index=3 split=dev done in 524.11s, precision=15.19 ± 4.21, recall=24.53 ± 3.30, f1=18.51 ± 3.58
[agent_gradient_eval_debug] prompt full evaluation: generation_index=3 beam_index=3 stable_f1=11.35 (k=2)
[agent_gradient_eval_debug] prompt full evaluation: generation_index=4 beam_index=4 split=dev done in 524.35s, precision=9.53 ± 2.04, recall=31.94 ± 12.99, f1=14.49 ± 3.44
[agent_gradient_eval_debug] prompt full evaluation: generation_index=4 beam_index=4 stable_f1=7.60 (k=2)"""

pattern = re.compile(r"f1=(\d+(?:\.\d+)?) ± (\d+(?:\.\d+)?)")

rows = []
for i, match in enumerate(pattern.finditer(text), start=1):
    mean = float(match.group(1))
    std = float(match.group(2))
    score = mean - 2.0 * std

    rows.append({
        "original_rank": i,
        "f1_mean": mean,
        "f1_std": std,
        "score": score,
    })

ranked = sorted(rows, key=lambda x: x["score"], reverse=True)

for rank, row in enumerate(ranked, start=1):
    print(
        f"{rank:2d}. line={row['original_rank']:2d} "
        f"f1={row['f1_mean']:.2f} ± {row['f1_std']:.2f} "
        f"score={row['score']:.2f}"
    )
```
