```python
import re

text = """[agent_gradient_eval_debug] prompt full evaluation: generation_index=0 beam_index=0 split=dev done in 955.03s, precision=1336.81 ± 447.43, recall=3279.09 ± 636.97, f1=1866.56 ± 488.80
[agent_gradient_eval_debug] prompt full evaluation: generation_index=1 beam_index=1 split=dev done in 992.72s, precision=1304.68 ± 271.05, recall=3782.80 ± 913.31, f1=1911.81 ± 321.32
[agent_gradient_eval_debug] prompt full evaluation: generation_index=2 beam_index=2 split=dev done in 992.89s, precision=1368.73 ± 400.99, recall=3464.28 ± 836.76, f1=1922.55 ± 441.37
[agent_gradient_eval_debug] prompt full evaluation: generation_index=3 beam_index=3 split=dev done in 993.69s, precision=1287.65 ± 316.52, recall=3464.28 ± 836.76, f1=1851.85 ± 382.36
[agent_gradient_eval_debug] prompt full evaluation: generation_index=4 beam_index=4 split=dev done in 991.12s, precision=1322.02 ± 302.52, recall=3597.61 ± 672.24, f1=1905.99 ± 330.80"""

pattern = re.compile(r"f1=(\d+(?:\.\d+)?) ± (\d+(?:\.\d+)?)")

rows = []
for i, match in enumerate(pattern.finditer(text), start=1):
    mean = float(match.group(1))
    std = float(match.group(2))
    score = mean/100 - 2.5 * std/100

    rows.append({
        "original_rank": i,
        "f1_mean": mean,
        "f1_std": std,
        "score": score,
    })

ranked = sorted(rows, key=lambda x: x["score"], reverse=True)

for rank, row in enumerate(ranked, start=1):
    print(
        f"{rank:2d}. line={row['original_rank']:2d} "
        f"f1={row['f1_mean']:.2f}±{row['f1_std']:.2f} "
        f"score={row['score']:.3f}"
    )
```
